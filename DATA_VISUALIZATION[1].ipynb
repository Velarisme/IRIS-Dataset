```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a bar plot with pink color
plt.bar(months, energy_consumption, color='pink')

# Add titles and Labels
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for t (from 0 to 2pi)
t = np.linspace(0, 2 * np.pi, 1000)

# Parametric equations for the heart shape
x = 16 * np.sin(t)**3
y = 13 * np.cos(t) - 5 * np.cos(2*t) - 2 * np.cos(3*t) - np.cos(4*t)

# Create the heart-shaped plot
plt.plot(x, y, color='red')

# Add title and labels
plt.title('Heart Shape Plot')
plt.xlabel('X')
plt.ylabel('Y')

# Display the plot
plt.axis('equal')  # Ensures the aspect ratio is equal
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Create a pie chart
plt.pie(energy_consumption, labels=months, autopct='%1.1f%%', colors=['pink', 'lightblue', 'lightgreen', 'lightcoral', 'lightyellow', 'lightskyblue'])

# Add title
plt.title('Energy Consumption by Month')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
energy_consumption = [1200, 1300, 1100, 1500, 1400, 1600]

# Convert months to numerical values for plotting (1 for Jan, 2 for Feb, etc.)
month_numbers = [1, 2, 3, 4, 5, 6]

# Create a scatter plot
plt.scatter(month_numbers, energy_consumption, color='pink')

# Add titles and Labels
plt.title('Energy Consumption Over 6 Months')
plt.xlabel('Month')
plt.ylabel('Energy Consumption (MWh)')
plt.xticks(month_numbers, months)  # Show month names on the x-axis

# Display the plot
plt.show()
```
